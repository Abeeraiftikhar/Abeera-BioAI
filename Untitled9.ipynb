```python
from groq import Groq

client = Groq(api_key=os.environ.get("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"))

def chat_with_groq(message, history):
    messages = [{"role": "system", "content": "You are a helpful chatbot."}]

    # Add previous conversation (memory)
    for user_msg, bot_msg in history:
        messages.append({"role": "user", "content": user_msg})
        messages.append({"role": "assistant", "content": bot_msg})

    # Add current message
    messages.append({"role": "user", "content": message})

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
    )

    return response.choices[0].message.content
```

```python
import gradio as gr

demo = gr.ChatInterface(
    fn=chat_with_groq,
 title="Abeera BioAI 🧬",
description="An AI-powered bioinformatics assistant for sequence analysis and computational biology."
)

demo.launch()
```
